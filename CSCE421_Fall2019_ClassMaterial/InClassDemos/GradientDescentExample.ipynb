```
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def  cost_function(x):
   '''
   x = input value
   Returns the output of f(x)=x^4+4x^3-2x^2+12x-3
   '''
   return math.pow(x,4)-4*math.pow(x,3)-2*math.pow(x,2)+12*x-3
```

```
# x_array=list(range(-5,5,0.001))
x_array=np.arange(-3,4, 0.1)
f_array = np.zeros(len(x_array))
i=0
for it in x_array:
    f_array[i]  = cost_function(it)
    i=i+1

fig,ax = plt.subplots(figsize=(6,4))
ax.set_ylabel('f(x)')
ax.set_xlabel('x')
_=ax.plot(x_array,f_array,'b.')
# _=ax.plot(x_array,x_array,'b.')

print(cost_function(1))
```

```
def  cost_function_1stder(x):
   '''
   x = input value
   Returns the output of f(x)=x^4+4x^3-2x^2+12x-3
   '''
   return 4*math.pow(x,3)-12*math.pow(x,2)-4*x+12
```

```
def gradient_descent(x_0,learning_rate=0.01,iterations=20):
    '''    
    Returns the history of cost and solution over no of iterations
    '''
    x_history = np.zeros(iterations)
    f_history = np.zeros(iterations)
    x=x_0;
    for it in range(iterations):
        x=x-learning_rate*cost_function_1stder(x)
        f_history[it]  = cost_function(x)
        x_history[it]  = x
        
    return x_history,f_history
```

```
iterations=100

x0=2
learning_rate=0.01 # converging, x0=2
# learning_rate=0.001 # almost converging, x0=2
# learning_rate=0.07 # oscillations, x0=2

#x0=0
#learning_rate=0.01 # converging, x0=0
#learning_rate=0.001 # almost converging, x0=0
#learning_rate=0.07 # oscillations, x0=0

x_history,f_history=gradient_descent(x0,learning_rate,iterations)
print(x_history)

fig,ax = plt.subplots(figsize=(6,4))
ax.set_ylabel('f(x)')
ax.set_xlabel('Iterations')
_=ax.plot(range(100),f_history,'b.')

fig,ax = plt.subplots(figsize=(6,4))
ax.set_ylabel('Iterations')
ax.set_xlabel('x')
_=ax.plot(x_history,range(100),'b.')
```
